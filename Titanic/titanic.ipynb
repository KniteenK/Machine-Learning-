```python
data = pd.read_csv (r'C:\\Users\\Kumar\Downloads\\git uploads\\Machine-Learning-\\Titanic\\train.csv')
test = pd.read_csv(r'C:\\Users\\Kumar\Downloads\\git uploads\\Machine-Learning-\\Titanic\\test.csv')
data.sample(5)
```

```python
data.shape
```

```python
data = data.drop(['PassengerId' , 'Name' , 'Ticket' , 'Fare' , 'Cabin'] , axis = 'columns')
submission = test[['PassengerId']]
test = test.drop(['PassengerId' , 'Name' , 'Ticket' , 'Fare' , 'Cabin'] , axis = 'columns')
```

```python
data.sample(3)
```

```python
data['family'] = data['SibSp'] + data['Parch']
data = data.drop(['SibSp' , 'Parch'] , axis = 1)
test['family'] = test['SibSp'] + test['Parch']
test = test.drop(['SibSp' , 'Parch'] , axis = 1)
data.head()
```

```python
test.head()
```

```python
data.head()
```

```python
data['Sex'] = data['Sex'].map({'male' : 1 , 'female' : 0})
data['Embarked'] = data['Embarked'].map({'S' : 0 , 'Q' : 1 , 'C' : 2})
test['Sex'] = test['Sex'].map({'male' : 1 , 'female' : 0})
test['Embarked'] = test['Embarked'].map({'S' : 0 , 'Q' : 1 , 'C' : 2})
data.head()
```

```python
test.head()
```

```python
data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x = data.iloc[: , 1 : ]
y = data.iloc[: , 0]
x_train , x_test , y_train , y_test = train_test_split (x , y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.impute import SimpleImputer 
simpleImputer = SimpleImputer()
embarked_imputer = SimpleImputer(strategy = 'most_frequent')
```

```python
x_train['Age'] = simpleImputer.fit_transform (x_train[['Age']])
x_test['Age'] = simpleImputer.transform (x_test[['Age']])
test['Age'] = simpleImputer.transform(test[['Age']])
```

```python
x_train['Embarked'] = embarked_imputer.fit_transform(x_train[['Embarked']]).ravel()
x_test['Embarked'] = embarked_imputer.transform(x_test[['Embarked']]).ravel()
test['Embarked'] = embarked_imputer.transform(test[['Embarked']]).ravel()
```

```python
x_test.isnull().sum()
```

```python
data.sample()
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_pred , y_test)
```
